```python
import pandas
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pandas.read_csv('../data/clean_data.csv')
```

```python
import gensim
from gensim.models.doc2vec import TaggedDocument
```

```python
train, test = train_test_split(data.clean_text, random_state=42, test_size=0.2)
```

```python
def handle_format(text_list, train=True):
    output = []
    for index, value in enumerate(text_list):
        tag = "train_{0}".format(index) if train else "test_{0}".format(index)
        output.append(TaggedDocument(value.split(), [tag]))
    return output
```

```python
train_formatted = handle_format(train)
test_formatted = handle_format(test, False)
```

```python
model_data = train_formatted + test_formatted
```

```python
model = gensim.models.Doc2Vec.load('../models/doc2vec/nassai_dbow_doc2vec.vec')
```

```python
doc_id = np.random.randint(model.docvecs.count)  # Pick random doc; re-run cell for more examples
print('for doc %d...' % doc_id)
inferred_docvec = model.infer_vector(model_data[doc_id].words)
print('%s:\n %s' % (model, model.docvecs.most_similar([inferred_docvec], topn=3)))
```
